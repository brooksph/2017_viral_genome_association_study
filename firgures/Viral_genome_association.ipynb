```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('all.csv', header = 0)
df
```

```python
#Ploting frequency of detection for each virus 

df['name'].value_counts().plot(kind="barh")
plt.savefig('viral_genome_hits.png', bbox_inches='tight')
```

```python
#Ploting genome fractions for each virus detected
newdf = df[df.columns[0:2]]
newdf_byname = newdf.set_index('name')
newdf_byname.groupby(level=0).mean().plot(kind="barh")
plt.savefig('viral_genome_fraction.png', bbox_inches='tight')
```
